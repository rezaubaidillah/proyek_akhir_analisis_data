## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
year_to_show = 2011  # Replace with the desired year

# Filter the data for the chosen year before resampling
filtered_df = day_df[day_df['dteday'].dt.year == year_to_show]

monthly_orders_df = filtered_df.resample(rule='m', on='dteday').agg({
    "instant": "nunique",
    "cnt": "sum"
})
monthly_orders_df.index = monthly_orders_df.index.strftime('%m') #mengubah format order date menjadi nama bulan

monthly_orders_df = monthly_orders_df.reset_index()
monthly_orders_df.rename(columns={
    "instant": "order_count",
    "cnt": "count"
}, inplace=True)


plt.figure(figsize=(10, 5))
plt.plot(monthly_orders_df["dteday"], monthly_orders_df["count"], marker='o', linewidth=2, color="#72BCD4")
plt.title("Number of Orders per Month (2011)", loc="center", fontsize=20)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```python
year_to_show = 2012  # Replace with the desired year

# Filter the data for the chosen year before resampling
filtered_df = day_df[day_df['dteday'].dt.year == year_to_show]

monthly_orders_df = filtered_df.resample(rule='m', on='dteday').agg({
    "instant": "nunique",
    "cnt": "sum"
})
monthly_orders_df.index = monthly_orders_df.index.strftime('%m') #mengubah format order date menjadi nama bulan

monthly_orders_df = monthly_orders_df.reset_index()
monthly_orders_df.rename(columns={
    "instant": "order_count",
    "cnt": "count"
}, inplace=True)


plt.figure(figsize=(10, 5))
plt.plot(monthly_orders_df["dteday"], monthly_orders_df["count"], marker='o', linewidth=2, color="#72BCD4")
plt.title("Number of Orders per Month (2012)", loc="center", fontsize=20)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

Berdasarkan visualisasi di atas, kita dapat melihat bahwa jumlah cnt(total perental) terbanyak terjadi pada bulan 9. Selain itu, kita juga dapat melihat adanya penurunan jumlah rental yang cukup signifikan pada bulan 10, 11, dan 12

### Pertanyaan 2:

```python
hourly_mean_cnt = hour_df.groupby('hr')['cnt'].sum().sort_values(ascending=False)

# Plot bar chart
plt.figure(figsize=(10, 6))
hourly_mean_cnt.plot(kind='bar')
plt.title(' jumlah perental berdasarkan waktu')
plt.xlabel('Jam')
plt.ylabel('Jumlah')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

Berdasarkan visualisasi di atas, kita dapat melihat bahwa jumlah cnt(total perental) terbanyak terjadi di jam 17 dan kita bisa melihat pada jam 1-5 kebawah hanya sedikit yang merental

## Conclusion

- setiap tahun pada bulan 5-10 memiliki high demand untuk merental dan memiliki low demand pada bulan 1-4 dan 11-12
